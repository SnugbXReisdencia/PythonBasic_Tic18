```python
import os

ARQUIVO_TAREFAS = 'tarefas.txt'

def carregar_tarefas():
    tarefas = []
    if os.path.exists(ARQUIVO_TAREFAS):
        with open(ARQUIVO_TAREFAS, 'r') as arquivo:
            for linha in arquivo:
                id_tarefa, descricao, status = linha.strip().split(',')
                tarefas.append({'id': int(id_tarefa), 'descricao': descricao, 'status': status})
    return tarefas

def salvar_tarefas(tarefas):
    with open(ARQUIVO_TAREFAS, 'w') as arquivo:
        for tarefa in tarefas:
            linha = f"{tarefa['id']},{tarefa['descricao']},{tarefa['status']}\n"
            arquivo.write(linha)

def listar_tarefas(tarefas):
    print("Lista de Tarefas:")
    for tarefa in tarefas:
        print(f"{tarefa['id']}. {tarefa['descricao']} {tarefa['status']}")

def registrar_tarefa(tarefas):
    descricao = input("Digite a descrição da tarefa: ")
    descricao = descricao.capitalize()
    nova_tarefa = {'id': len(tarefas) + 1, 'descricao': descricao, 'status': '[ ]'}
    tarefas.append(nova_tarefa)
    print("Tarefa registrada!!!")
    salvar_tarefas(tarefas)

def marcar_como_realizada(tarefas):
    listar_tarefas(tarefas)
    id_tarefa = int(input("Digite o ID da tarefa a ser marcada como realizada: "))
    
    for tarefa in tarefas:
        if tarefa['id'] == id_tarefa and tarefa['status'] == '[ ]':
            tarefa['status'] = '[x]'
            tarefas.remove(tarefa)
            tarefas.insert(0, tarefa)
            print("Tarefa marcada como realizada!")
            salvar_tarefas(tarefas)
            return
    
    print("Tarefa não encontrada ou já realizada.")

def editar_tarefa(tarefas):
    listar_tarefas(tarefas)
    id_tarefa = int(input("Digite o ID da tarefa a ser editada: "))
    
    for tarefa in tarefas:
        if tarefa['id'] == id_tarefa:
            nova_descricao = input("Digite a nova descrição da tarefa: ")
            tarefa['descricao'] = nova_descricao.capitalize()
            print("Tarefa editada!")
            salvar_tarefas(tarefas)
            return
    
    print("Tarefa não encontrada.")

# Main
tarefas = carregar_tarefas()

while True:
    print("\n====== ToDoList ======")
    print("1. Listar Tarefas")
    print("2. Registrar Tarefa")
    print("3. Marcar como Realizada")
    print("4. Editar Tarefa")
    print("0. Sair")
    
    opcao = input("Escolha uma opção: ")
    
    if opcao == '1':
        listar_tarefas(tarefas)
    elif opcao == '2':
        registrar_tarefa(tarefas)
    elif opcao == '3':
        marcar_como_realizada(tarefas)
    elif opcao == '4':
        editar_tarefa(tarefas)
    elif opcao == '0':
        print("Saindo do aplicativo ToDoList. Até mais!")
        break
    else:
        print("Opção inválida. Tente novamente.")
```
